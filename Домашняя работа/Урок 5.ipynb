Отфильтрованные данные записаны в retail. Идентификатор заказа – InvoiceNo. Для каждого заказа в данных может встречаться более 1 строки.

Подсказки

- **Как решать это задание?**
    
    Отберите клиентов только из Германии (`Country`), сгруппируйте по клиентам (`CustomerID`) и посчитайте число заказов для каждого клиента. Затем на полученном результате используйте метод `quantile`, чтобы получить данные по 80-ому перцентилю. После этого отфильтруйте Германских клиентов, оставив с числом заказов больше чем 80-ый перцентиль. Запишите `CustomerID` полученных клиентов в переменную `germany_top`
    
- **У меня не сходится ответ!**
    
    А какую функцию вы применяли при агрегации после группировки по клиентам? Там нужно посчитать число заказов, при том, что на 1 заказ может быть несколько строк - например, заказ получен, заказ в процессе, заказ доставлен. Поэтому в агрегации нужно использовать функцию `'nunique'` c колонкой `InvoiceNo`.
    
- **Подсчёт числа уникальных значений**
    
    Используйте метод `nunique`, чтобы получить число уникальных значений для каждой группы. В агрегации его можно указать как метод `pd.Series.nunique` либо как строку `'nunique'` (такой вариант работает быстрее)
    
- **Как взять 80-ый перцентиль?**
    
    Воспользуйтесь методом `quantile`, указав внутри него какой перцентиль вам нужен в виде дроби. К примеру
    
    ```python
    df.column1.quantile(0.5)
    ```
    
    даст вам 50-ый перцентиль по значениям в колонке `column1`
    
- **Как использовать переменную внутри** `query`**?**
    
    Чтобы сослаться на переменную внутри `query`, напишите её в запросе и поставьте перед ней `@`. Например
    
    ```python
    threshold = 3
    df.query('col1 >= @threshold')
    ```
    
    отберёт из датафрэйма строки, где в колонке `col1` находятся значения больше либо равные 3
    
- **Я использовал** `groupby`**, не указывая** `as_index=False`**, и** `CustomerID` **оказались в индексе, как их достать?**
    
    Самый простой вариант - обратитесь к атрибуту `index` у нужного датафрэйма. То есть
    
    ```python
    df.index
    ```
    
    Вернёт вам индекс датафрэйма `df`

````python
count_germany=q_iso.query('Country == "Germany"')\
                    .groupby('CustomerID', as_index=False)\
                    .agg({'InvoiceNo' : 'nunique'})
quant=count_germany.InvoiceNo.quantile(q=0.8)
count_germany=count_germany.query('InvoiceNo > @quant')
germany_top=count_germany.CustomerID
````

Вернемся к анализу полного датасета retail. Создайте колонку Revenue с суммой покупки, используя колонки Quantity и UnitPrice.

- **Как посчитать сумму покупки?**
    
    Для этого нужно умножить количество купленного товара на его цену, например
    
    Коржики стоят по 7 рублей
    
    Я взял 8 коржиков
    
    Сумма покупки будет
    
    8 штук * 7 рублей == 56 рублей
    
- **Создание колонки**
    
    Чтобы сделать новую колонку, достаточно написать такой код
    
    ```python
    df['new_column'] = some_value
    ```
    
    где
    
    `df` - датафрэйм
    
    `new_column` - название новой колонки
    
    `some_value` - то, чему будет равна колонка (это может быть одно число или колонка датафрэйма, где число элементов такое же как число строк в `df`)

````python
q_iso[['Quantity', 'UnitPrice']] = q_iso[['Quantity', 'UnitPrice']]. fillna (0)
````

````python
q_iso.assign(ctr=q_iso.Quantity * q_iso.UnitPrice)
````
