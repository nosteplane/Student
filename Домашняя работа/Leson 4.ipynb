Посмотрим, какие водители опаздывали на заказы к определенному времени (start_type == 'reserved'). Опозданием считается время ожидания wait_time > 0.0.

Сгруппируйте отфильтрованные данные по driver_id и в качестве ответа укажите id водителя, который опоздал на заказ интересующего нас типа наибольшее число раз

- **Как решать эту задачу?**
    
    Для начала отфильтруем строки, оставив те, где `start_type` `reserved` и время ожидания больше 0 (то есть водитель опоздал и клиент ждал). Далее нужно сгруппировать данные по водителям (`driver_id`), посчитать число опозданий и выбрать водителя с максимальным числом опозданий
    
- **Как отобрать времена больше 0?**
    
    Просто укажите в `query`, что колонка со временем больше 0
    
- **Отбор индекса с максимальным значением**
    
    Индекс максимального значения можно достать методом `idxmax` . 
    
    **Важный момент!** Чтобы использовать `idxmax`, нужно, чтобы айдишник водителя был в индексах. Это тот самый случай, когда при группировке **не нужно** использовать `as_index=False`.

````python
taxi_res = taxi.query('start_type == "reserved"') 
taxi_res_taim = taxi_res.query('wait_time > 0.0')
taxi.groupby('driver_id').agg({'taxi_res_taim':'count'}).idxmax()
````

Создайте колонку wait_time, в которой будет храниться разница между временем подачи машины (arrived_at) и временем заказа (start_at) в минутах.

Чтобы сконвертировать полученный результат в минуты, можно использовать метод  .astype('timedelta64[m]')

- **А что делать тут?**
    
    Здесь создадим новую колонку `wait_time` в датафрэйме, куда поместим разницу между временами `arrived_at` и `start_at`. А так же переведём эту разницу в минуты
    
- **Как посчитать разницу между временами?**
    
    Просто вычтите из одной колонке временного типа другую такую колонку. Это даст колонку типа `timedelta`. Например,
    
    ```python
    df.time1 - df.time2
    ```
    
    Вернёт колонку с разницей между соответствующими ячейками колонки `time1` и `time2`
    
- **Как сконвертировать** `timedelta` **в минуты?**
    
    Используйте метод astype, передав в него тип `'timedelta64[m]'`. К примеру,
    
    ```python
    df.time_column.astype('timedelta64[m]')
    ```
    
    Вернёт колонку time_column, переведённую в минуты

````python
taxi['wait_time'] = None 
taxi['wait_time'] = taxi.arrived_at - taxi.start_at
taxi['wait_time'] = taxi.wait_time.astype('timedelta64[m]')
````
